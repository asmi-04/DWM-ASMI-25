**AIM** : Implementation of Association Rule Mining Algorithm (Apriori)

**INTRODUCTION** : Association Rule Mining is one of the most important techniques in data mining used to discover interesting relationships, patterns, and correlations among sets of items in transactional databases. The Apriori algorithm, developed by R. Agrawal and R. Srikant, is a fundamental algorithm used for mining frequent itemsets and deriving association rules. It works on the principle that if an itemset is frequent, then all of its subsets must also be frequent. This algorithm is widely used in market basket analysis to find combinations of items that frequently co-occur in transactions.

**PROCEDURE:**
1. Dataset Preparation: Create a small transactional dataset containing item purchases per customer.
2. Import Required Libraries: Use mlxtend library for Apriori and association rules.
3. Convert Transactions to DataFrame.
4. Apply Apriori Algorithm: Generate frequent itemsets with minimum support threshold.
5. Generate Association Rules: Derive rules with minimum confidence threshold.
6. Analyze Output:
Examine rules like-
If bread → milk (support: 0.6, confidence: 0.75)
If milk & bread → butter (confidence: 0.75)

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import networkx as nx
import matplotlib.pyplot as plt
data = {
'Bread': [1, 1, 0, 1, 1],
'Milk': [1, 1, 1, 1, 0],
'Butter': [0, 1, 1, 1, 1],
'Cheese': [1, 0, 1, 1, 1]
}

df = pd.DataFrame(data)
print("Transaction Data:")
print(df)
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
print("\nFrequent Itemsets:")
print(frequent_itemsets)
rules = association_rules(frequent_itemsets, metric="confidence",
min_threshold=0.7)
print("\nAssociation Rules:")
print(rules)
G = nx.DiGraph()
for index, row in rules.iterrows():
  antecedent = list(row['antecedents'])[0]
  consequent = list(row['consequents'])[0]
  G.add_edge(antecedent, consequent, weight=row['confidence'])
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G, k=0.5, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue',
font_size=12, font_weight='bold', edge_color='gray')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Association Rules Graph")
plt.show()
```

**CONCLUSION** :

The Apriori algorithm successfully discovered meaningful associations from the transactional dataset. From the generated rules, we observed that items like milk and bread frequently occur together and often lead to the purchase of butter. This experiment demonstrates the power of association rule mining in uncovering hidden patterns in customer purchase behavior. Such insights can be used by businesses for targeted marketing, inventory management, and sales promotions. The Apriori algorithm, despite its simplicity, remains a powerful tool in the field of data mining.

1. What is the Apriori algorithm in Association Rule Mining?

  Ans: The Apriori algorithm is a fundamental technique in Association Rule Mining used to
identify frequent itemsets in large datasets and derive strong association rules. It follows a
bottom-up approach, where frequent itemsets are extended one item at a time (the Apriori
property), ensuring that all subsets of a frequent itemset are also frequent. The algorithm
consists of two main steps: candidate generation (finding potential frequent itemsets) and
pruning (removing infrequent ones). It is widely used in market basket analysis,
recommendation systems, and data mining applications to uncover relationships between
items, such as products frequently purchased together.

---



2. What is the significance of support, confidence, and lift in Apriori?

  Ans: Support, confidence, and lift are key metrics in the Apriori algorithm. **Support** indicates how frequently an itemset appears in the dataset. **Confidence** shows how often item B appears in transactions that contain item A, reflecting the reliability of the rule. **Lift** measures the strength of a rule over random chance; a lift greater than 1 indicates a strong positive correlation. Together, these metrics help identify meaningful and useful association rules in transactional data.
